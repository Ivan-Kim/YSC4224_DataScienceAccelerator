### Exercise 1.0
Load the tables below as `DataFrames`. Assign them respectively to the variables `student_detail` and `exam_data`.

````python
student_id_list = ["S" + str(i) for i in range(1,14) if i not in [6,10,13]]
student_id = pd.Series(student_id_list)
name_list = ["Danniella Fenton", "Ryder Storey", "Bryce Jensen", "Ed Bernal", "Kwame Morin", "Mark Smith", "Bob Love", "Bruce Ong", "Alex Sander", "Allie Chang"]
name = pd.Series(name_list)
cumulative_marks_list = [200, 210, 190, 222, 199, 201, 214, 193, 230, 183]
cumulative_marks = pd.Series(cumulative_marks_list)
student_detail = pd.DataFrame({"student_id": student_id, "name": name, "cumulative_marks": cumulative_marks})
student_detail
````

````python
student_id_list = ["S" + str(i) for i in range(1,14) if i != 5]
student_id = pd.Series(student_id_list)
exam_score_list = [23, 45, 12, 67, 21, 55, 33, 14, 56, 83, 88, 12]
exam_score = pd.Series(exam_score_list)
exam_data = pd.DataFrame({"student_id": student_id, "exam_score": exam_score})
exam_data
````

### Exercise 1.1   

Include the exam score from `exam_data` for each student listed in the `student_detail` table. 


````python
new_student_detail = pd.merge(student_detail, exam_data, how="left")
new_student_detail
````


### Exercise 1.2
Fill any missing score with the median exam score. 


````python
missingRow = new_student_detail[new_student_detail["exam_score"].isnull()]
medianScore = new_student_detail["exam_score"].median()
print(medianScore)
# new_student_detail.loc[missingRow.index, "exam_score"] = medianScore
new_student_detail["exam_score"].fillna(medianScore, inplace=True)
new_student_detail
````


### Exercise 1.3
Calculate the final average score, considering that the `cumulative_marks` is the sum of 3 different exams. 



````python
new_student_detail["average"] = (new_student_detail["cumulative_marks"] + new_student_detail["exam_score"]) / 4
new_student_detail
````

### Exercise 1.4
Assign a letter grade to each student based on their relative performance (percentiles): top 25% gets an A, second 25% chunk gets a B and so on. 

````python
grades = pd.qcut(new_student_detail["average"], q = 4, labels = ["D", "C", "B", "A"])
new_student_detail["grades"] = grades
new_student_detail
````
